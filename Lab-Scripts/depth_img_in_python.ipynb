## Displaying Depth and RGB Image
The below code displays the depth and RGB image. The first function "load_and_display_image" just displays a depth image and the second function "load_and_display_two_images" first takes in the depth image and then the rgb image to display side by side. Because we are using "%matplotlib widget" above, we are able to hover over a pixel and get its pixel value which is useful for immediately seeing the relative depth.

```python
def load_and_display_image(image_path, fig_h=8, fig_w=4):
    # Load image using OpenCV in grayscale mode
    img_gray = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
    
    # Display image using Matplotlib
    fig, ax = plt.subplots(figsize=(fig_h, fig_w))
    ax.imshow(img_gray, cmap='gray')
    # ax.axis('off')  # Turn off axis
    
    plt.show()
```

```python
def load_and_display_two_images(image_path1, image_path2):
    # Load first image using OpenCV in grayscale mode
    img_gray = cv.imread(image_path1, cv.IMREAD_GRAYSCALE)
    
    # Load second image using OpenCV in color mode
    img_color = cv.imread(image_path2)
    img_color = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)
    # Display images side by side using Matplotlib
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    
    # Display first image (grayscale)
    axes[0].imshow(img_gray, cmap='gray')
    axes[0].set_title('Grayscale Image')
    axes[0].axis('off')  # Turn off axis
    
    # Display second image (color)
    axes[1].imshow(img_color)
    axes[1].set_title('Color Image')
    axes[1].axis('off')  # Turn off axis
    
    plt.show()
```

## Create Relative Depth Image
The below piece of code creates relative depth images from the images stored in the images/fabricated_images folder. In order to run it you need a conda environement called depth_anything where you have installed all of the dependencies in the ../Depth-Anything/requirements.txt file

```python
# You have to have a conda environment set up named depth_anything that follows the environment creation steps outlined in this repo:
# https://github.com/LiheYoung/Depth-Anything/tree/main 
def process_images(input_directory, output_directory):
    os.chdir("../Depth-Anything")
    conda_env = "depth_anything"
    # Activate the Conda environment
    activate_command = f"conda activate {conda_env}"
    subprocess.run(activate_command, shell=True)
    
    # Make sure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Get a list of all files in the input directory with specified extensions
    image_extensions = [".jpg", ".jpeg", ".png"]
    image_files = [file for file in os.listdir(input_directory) if os.path.isfile(os.path.join(input_directory, file)) and file.lower().endswith(tuple(image_extensions))]

    # Run the command for each image file
    for image_file in image_files:
        input_path = os.path.join(input_directory, image_file)
        output_path = os.path.join(output_directory)
        command = f"python ..\\Depth-Anything\\run.py --encoder vitl --img-path \"{input_path}\" --outdir \"{output_path}\" --pred-only --grayscale"
        subprocess.run(command, shell=True)
        print(output_path)

    # Deactivate the Conda environment
    deactivate_command = "conda deactivate"
    subprocess.run(deactivate_command, shell=True)

runDepthModel = False

if __name__ == "__main__" and runDepthModel:
    input_directory = "../Lab-Scripts/images/fabricated_images"
    output_directory = "../Lab-Scripts/images/relative_depth_images"
    process_images(input_directory, output_directory)
```
